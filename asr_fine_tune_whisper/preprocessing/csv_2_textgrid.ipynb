```python
#tiers = ['text', 'text', 'transcription', 'transcription', 'transcription_original', 'transcription_original']
#tiers_name = ['text-1', 'text-2', 'whisper-fine-tune-1', 'whisper-fine-tune-2', 'whisper-original-1', 'whisper-original-2']
#tiers_sub = [1,2,1,2,1,2]

#tiers = ['transcription', 'transcription', 'transcription_original', 'transcription_original']
#tiers_name = ['whisper-fine-tune-1', 'whisper-fine-tune-2', 'whisper-original-1', 'whisper-original-2']
#tiers_sub = [1,2,1,2]

#tiers = ['transcription', 'transcription', 'transcription_original', 'transcription_original', 'transcription_v3', 'transcription_v3']
#tiers_name = ['whisper-fine-tune-1', 'whisper-fine-tune-2', 'whisper-original-1', 'whisper-original-2', 'whisper-fine-tune-v3-1', 'whisper-fine-tune-v3-2']
#tiers_sub = [1,2,1,2,1,2]

tiers = ['text', 'text', 'transcription_v1', 'transcription_v1', 'transcription_v2', 'transcription_v2', 'transcription_v3', 'transcription_v3', 'transcription_original', 'transcription_original']
tiers_name = ['text-1', 'text-2', 'whisper-v1-1', 'whisper-v1-2', 'whisper-v2-1', 'whisper-v2-2', 'whisper-v3-1', 'whisper-v3-2', 'whisper-original-1', 'whisper-original-2']
tiers_sub = [1,2,1,2,1,2,1,2,1,2]


lines = []
lines_head = ['File type = "ooTextFile"',
'Object class = "TextGrid"', '']
xmin_all = 'xmin = ' + str(0)
xmax_all = 'xmax = ' + str(580) # change xmax



tier_info_general = ['tiers? <exists>', 
'size = 10', 'item []:'] # change tier number

sec_all = []
for i in range(10): # change tier number
    df_curr = df[df['file_tier'] == tiers_sub[i]]
    text_curr = df_curr[tiers[i]].tolist()
    text_curr = [str(item).replace('"', "'") for item in text_curr]
    xmin_list = df_curr['start'].tolist()
    xmax_list = df_curr['end'].tolist()

    
    print (tiers[i])
    tier_info = [ 
    '    item [' + str(i + 1) +']:',
    '        class = "IntervalTier"', 
    '        name = "'+ tiers_name[i] +'"'             
    ]
    interval = 'intervals: size = ' + str(len(text_curr))
    sec_all = sec_all + tier_info + [xmin_all,xmax_all] + [interval]
    for j in range(len(text_curr)):
        interval_curr = 'intervals [' + str(j + 1) + ']:'
        xmin = 'xmin = ' + str(xmin_list[j])
        xmax = 'xmax = ' + str(xmax_list[j])
        text = 'text = "' + str(text_curr[j]) + '"'
        sec_all = sec_all + [interval_curr, xmin, xmax, text]

lines = lines_head + [xmin_all,xmax_all] + tier_info_general + sec_all
```
