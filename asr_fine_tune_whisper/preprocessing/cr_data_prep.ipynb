```python
import pandas as pd
import os
cwd = os.getcwd()
cwd
```

```python

#tail = '_overlap_removed_30'
tail = '_10'
save_path = '/work/jh2lab/elic_model'
save_chunk_path = save_path + '/cr_audio_chunk' + tail + '/'
save_chunk_path
```

```python
folder_path = '/work/jh2lab/elic_model/cr_audio_data/cr_audio_1018'
```

```python
#df['audio_path'][0]
```

```python
df = pd.read_csv(cwd + '/cr_chunk_text_info' + tail + '.csv')
df
```

```python
# read in the audio, for each row in df_chunk, get the chunked audio and save
from pydub import AudioSegment

def get_audio_chunk(start, end, chunk_name, file_tier):
    audio_dir = folder_path +'/'+ file_tier[:-2] + '.wav'
    if '_overlap_removed_15' in audio_dir:
        audio_dir = audio_dir.replace('_overlap_removed_15', '')
        #chunk_name = chunk_name.replace('_overlap_removed_15', '')
    #print (audio_dir)
    t1 = start * 1000 # s to ms
    t2 = end * 1000
    newAudio = AudioSegment.from_wav(audio_dir)[t1:t2]
    newAudio.export(save_chunk_path + chunk_name + tail + '.wav', format="wav")
    return chunk_name + '.wav'
```

```python
df['audio_name'] = df.apply(lambda x: get_audio_chunk(x.start, x.end, x.chunk_name, x.file_tier), axis = 1)
df
```
