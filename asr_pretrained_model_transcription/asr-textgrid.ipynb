```python
import pandas as pd
df_word = pd.read_csv('/Users/irisz/downloads/2023_ELIC/ckm001-2022-01-16-Trviž_01_word.csv')
df_snt = pd.read_csv('/Users/irisz/downloads/2023_ELIC/ckm001-2022-01-16-Trviž_01_sentence.csv')
```

```python
df_word.fillna('', inplace=True)
df_snt.fillna('', inplace=True)
df_word
```

```python
tiers = ['text', 'classla', 'classla_lg', 'nvdia_text', 'whisper_text', 'whisper_pad']
tiers_name = ['text', 'classla', 'classla-lg', 'nvdia', 'whisper', 'whisper_pad']

lines = []
lines_head = ['File type = "ooTextFile"',
'Object class = "TextGrid"', '']
xmin_all = 'xmin = ' + str(0)
xmax_all = 'xmax = ' + str(175)

xmin_list = df_word['xmin'].tolist()
xmax_list = df_word['xmax'].tolist()

tier_info_general = ['tiers? <exists>', 
'size = 6', 'item []:']

sec_all = []
for i in range(6):
    text_curr = df_word[tiers[i]].tolist()
    print (tiers[i])
    tier_info = [ 
    '    item [' + str(i + 1) +']:',
    '        class = "IntervalTier"', 
    '        name = "'+ tiers_name[i] +'"'             
    ]
    interval = 'intervals: size = ' + str(len(text_curr))
    sec_all = sec_all + tier_info + [xmin_all,xmax_all] + [interval]
    for j in range(len(text_curr)):
        interval_curr = 'intervals [' + str(j + 1) + ']:'
        xmin = 'xmin = ' + str(xmin_list[j])
        xmax = 'xmax = ' + str(xmax_list[j])
        text = 'text = "' + str(text_curr[j]) + '"'
        sec_all = sec_all + [interval_curr, xmin, xmax, text]

lines = lines_head + [xmin_all,xmax_all] + tier_info_general + sec_all
```

```python
lines
```

```python
with open('/Users/irisz/downloads/2023_ELIC/' + 'ckm001-2022-01-16-Trviž_01_word.TextGrid', 'w') as f:
    for line in lines:
        f.write(f"{line}\n")
```
