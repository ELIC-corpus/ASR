```python
import re
import pandas as pd
import numpy as np



def get_audio_df(textgrid_dir):
    
    with open(textgrid_dir, 'r') as file:
        lines = [line.rstrip() for line in file]
       
    text_list = []
    xmin_list = []
    xmax_list = []
    interval_list = []
    for line in lines[14:]:  #informations needed begin on the 9th lines
        if 'text =' in line:
            line = line.split('"')[1]
            if line != "":
                text_list.append(line)
                #print (len(text_list), len(xmin_list), len(xmax_list), len(interval_list), line, interval_list[-2:])
            else:
                xmin_list = xmin_list[:-1]
                xmax_list = xmax_list[:-1]
                interval_list = interval_list[:-1]
        if 'xmin' in line:
            time_min = line.split('=')[1].strip(' ')
            xmin_list.append(time_min)
        if 'xmax' in line:
            time_max = line.split('=')[1].strip(' ')
            xmax_list.append(time_max)
        if 'intervals [' in line:
            interval = line.split('[')[1].strip(']:')
            interval_list.append(interval)
        if 'intervals: size' in line:
            xmin_list = xmin_list[:-1]
            xmax_list = xmax_list[:-1]
    xmin_list = [float(item) for item in xmin_list]
    xmax_list = [float(item) for item in xmax_list]
    interval_list = [int(item) for item in interval_list]

    df = pd.DataFrame()
    df['text'] = text_list
    df['xmin'] = xmin_list
    df['xmax'] = xmax_list
    df['interval'] = interval_list
    df['tier'] = get_interval(df['interval'].tolist())
    df['start'] = df['xmin'].apply(lambda x: int(np.floor(x*16000)))
    df['end'] = df['xmax'].apply(lambda x: int(np.ceil(x*16000)))
    return df

def get_interval(time_list):
    interval_all = []
    interval_curr = 1
    for i in range(len(time_list)):
        if i == 0:
            interval_all.append(1)
        else:
            if time_list[i] > time_list[i - 1]:
                interval_all.append(interval_curr)
                continue
            else:
                interval_curr = interval_curr + 1
                interval_all.append(interval_curr)
    return interval_all

```

```python
import numpy as np
from thefuzz import fuzz

def get_matched(t1, t2):
    #print (t1, t2)
    if t1 == t2:
            return 1
    else:
        return 0
def get_unmatched_cand(match_list, idx):
    cand = []
    idx_org = idx
    if match_list[idx] == 1:
        return [idx]
    else:
        while match_list[idx] != 1 and idx > 0:
            cand.append(idx)
            idx = idx - 1 # before
        idx = idx_org
        while match_list[idx] != 1 and idx < len(match_list) -1:
            cand.append(idx)
            idx = idx + 1 # after
        return list(set(cand))
def word_match_score(t1, t2):
    if t1 == '-' or t2 == '-':
        return 0
    else:
        score = fuzz.ratio(t1, t2)
        return score
def get_can_score(idx,match_cand):
    if len(match_cand) == 1:
        return [100]
    score_all = []
    for cand in match_cand:
        t1 = df['manual'][idx]
        t2 = df['model'][cand]
        score = word_match_score(t1, t2)
        score_all.append(score)
    return score_all
def get_fuzzy_cand(match_cand, match_cand_score):
    max_cand = np.max(match_cand_score)
    if max_cand > 60:
        # fuzzy matched
        max_cand_id = match_cand[match_cand_score.index(max_cand)]
        #print (max_cand_id)
        word = df['model'][max_cand_id]
        return word, max_cand
    else:
        return '-', 0
    
    
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
import pandas as pd

wer = {}
string1 = text_manual

text_list = [text_classla, text_classla_lg, text_nvdia, text_whisper]
model_list = ['classla', 'classla_lg', 'nvdia', 'whisper']

for i in range(4):
    string2 = text_list[i]
    alignments = pairwise2.align.globalxx(string1.split(), 
                                          string2.split(),
                                          gap_char=['-']
                                         )


    df = pd.DataFrame(columns = ['manual', 'model'])
    df['manual'] = alignments[0].seqA
    df['model'] = alignments[0].seqB
    df['idx'] = df.index    
    df['match'] = df.apply(lambda x:get_matched(x.manual, x.model), axis = 1)
    df['match_cand'] = df.apply(lambda x: get_unmatched_cand(df['match'].tolist(), x.idx), axis = 1)
    df['match_cand_score'] = df.apply(lambda x: get_can_score(x.idx, x.match_cand), axis = 1)
    df['model_fuzzy'] = df.apply(lambda x: get_fuzzy_cand(x.match_cand, x.match_cand_score)[0], axis = 1)
    df['model_fuzzy_score'] = df.apply(lambda x: get_fuzzy_cand(x.match_cand, x.match_cand_score)[1], axis = 1)
    
    # get WER here
    wer[model_list[i]] = get_wer(df['manual'].tolist(), df['model'].tolist())  #wer,s,d,i,c,n

    df_match = df[df['manual']!= '-']
    
    df_word[model_list[i] + '_fuzzy'] = df_match['model_fuzzy'].tolist()
    df_word[model_list[i] + '_fuzzy_score'] = df_match['model_fuzzy_score'].tolist()
```

```python
def get_wer(manual, model):
    s = 0 # substitution
    d = 0 # deletion
    i = 0 # insertion
    c = 0 # correct 
    n = 0 # s + d + c
    for i in range(len(manual)):
        t1 = manual[i]
        t2 = model[i]
        if (t1 != '-') and (t2 != '-') and (t1 != t2):
            s = s + 1
        if (t1 != '-') and (t2 == '-'):
            d = d + 1
        if (t1 == '-') and (t2 != '-'):
            i = i + 1
        if (t1 != '-') and (t2 != '-') and (t1 == t2):
            c = c + 1
    n = s + d + c
    wer = (s + d + i)/n * 100
    return wer,s,d,i,c,n
get_wer(df['manual'].tolist(), df['model'].tolist())    

# WER can be larger than 1
```
