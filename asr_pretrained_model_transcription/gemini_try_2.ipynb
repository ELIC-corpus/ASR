```python
# https://pypi.org/project/SpeechRecognition/

import speech_recognition as sr

#audio_seg = "/users/irisz/downloads/2023_elic/ckm009-2023-05-26-Žminj/seg_1.wav"
#audio_dir = "/users/irisz/downloads/2023_elic/ckm009-2023-05-26-Žminj/ckm009-2023-05-26-Žminj_02.wav"
audio_dir = "/users/irisz/downloads/2023_elic/ckm008-2023-05-29-Crikvenica/ckm008-2023-05-29-Crikvenica_01.wav"

# Create an instance of the Recognizer class
recognizer = sr.Recognizer()

# Create audio file instance from the original file
audio_ex = sr.AudioFile(audio_dir)
type(audio_ex)

```

```python
# Create audio data
with audio_ex as source:
    audiodata = recognizer.record(audio_ex, duration=6)
type(audiodata)
```

```python
# Extract text
text = recognizer.recognize_google(audio_data=audiodata, language='hr')

print(text)
```
