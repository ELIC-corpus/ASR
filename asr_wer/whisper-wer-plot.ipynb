```python
import pandas as pd

df_2 = pd.read_csv('/users/irisz/downloads/nlp_elic/whisper_wer_all.csv')
df_2
```

```python
interest_info = 'wer'
cols = []
for i in range(1, 11):
    col = 'transcription_v' + str(int(i)) + '_wer_' + interest_info
    cols.append(col)
cols
```

```python
import numpy as np
df_3 = df_2[cols]
df_4 = df_3.replace([np.inf, -np.inf], np.nan).dropna()
df_4
```

```python
cols_2 = ['whisper_' + x.split('_')[1] for x in cols]
cols_2[-1] = "whisper_large_v3"
cols_2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = sns.boxplot(x="variable", y="value", data=pd.melt(df_4), color='skyblue')
ax.set_xticklabels(cols_2)
ax.tick_params(labelsize=20)
ax.set_xlabel("Whisper Model",fontsize=30)
ax.set_ylabel("WER(%)",fontsize=30)
sns.set_theme(rc={'figure.figsize':(25, 10)}, style="darkgrid")
#sns.set(font_scale = 1)
plt.ylim(0, 300)
plt.xticks(rotation=30)
plt.show()
#ax.figure.savefig('/home/sz38235/workDir/elic/pic/whisper_wer.png',dpi=600, bbox_inches="tight")
```
